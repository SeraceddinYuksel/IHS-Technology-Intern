```python
Face detection was made with the help of MTCNN. The face was marked using the bounding box.
Bounding box is an important component that helps distinguish the face from other objects and reduces the face to a smaller size.
```

```python
import os  # Import the library that allows us to access photos from our computer.
import cv2  # Load the OpenCV library, commonly used in many image processing projects.
from facenet_pytorch import MTCNN  # Load the MTCNN library based on PyTorch for face detection.
import matplotlib.pyplot as plt  # Used for data visualization.
```

```python
# Showing image path
image_dir = r'C:\Users\SERACEDDIN\Desktop\github'
```

```python
# We create a list containing image files.
image_paths = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.jpg')]
```

```python
# Load the MTCNN
mtcnn = MTCNN()
```

```python

# We detect faces for all images and put those faces in a red square
for path in image_paths:
    # Read the image
    image = cv2.imread(path)
    
    # we give a warning if the image cannot be loaded
    if image is None:
        print(f"Error loading image: {path}")
        continue
    
    # Face detection
    boxes, _ = mtcnn.detect(image)
    
    # We frame the faces and visualize them.
    if boxes is not None:
        for box in boxes:
            x1, y1, x2, y2 = box.astype(int)
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    # We convert the image in BGR format to RGB format
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Showing the new image with matplotlib
    plt.figure(figsize=(10, 10))
    plt.imshow(image_rgb)
    plt.axis('off')  
    plt.show()
```

```python

This code reads `.jpg` image files from a specified directory, detects faces in each image, 
and highlights them with red rectangles. It then converts each image to the RGB color format 
and visualizes the results using matplotlib. This enables easy visual inspection of face detection operations.
```
